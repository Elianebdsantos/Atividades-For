```python
#4 Inicalize uma lista de 10 numeros interios. Armazene os números pares em uma lista PAR e os números impares em uma lista IMPAR. Imprema as listas PAR e IMPAR

lista_10=[1,5,7,6,8,9,10,20,18,32]
apenas_pares=[x for x in lista_10 if x % 2 == 0]
apenas_impares=[x for x in lista_10 if x % 2 != 0]
print(apenas_pares)
print(apenas_impares)
```

```python
#8. Em uma competição de ginástica, cada atleta recebe votos de sete jurados. A melhor e a pior nota são eliminadas.
#A sua nota fica  sendo a média dos votos restantes. Você deve fazer um programa que receba o nome do ginasta e as notas dos sete jurados alcançadas pelo atleta em sua apresentação e depois informe a sua média, conforme a descrição acima informada (retirar o melhor e o pior salto e depois calcular a média com as notas restantes). As notas não são informadas em ordem crescente ou decrescente


nome= input("Digite o nome do ginasta: ")
notas= []
cont = 0
n_jurados= 7
while cont < n_jurados:
  nota = float(input(f'nota{cont+1}:'))
  notas.append(nota)
  cont += 1
  print(notas)
  notas.sort()
  print(notas)
  notas_finais = notas [1:-1]
  print(notas_finais)
  media = sum(notas_finais)/len(notas_finais)
  print(f'Ginasta{nome}:nota{media}')
```
