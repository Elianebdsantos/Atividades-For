```python
# 1 Outra forma Exercício com a função do for e while
#Solicitar um numero inteiro positivo N ao usuário. Imprimir os numeros de N a 1. Realizar o exercicio usando loops for e while
N= int(input("Digite um numero interio maior que 1"))
print("For:")
for n in range (N):
  print(f"{N-n}", end=" ")
print("\n")
print("while:")
i= 0

while i < N:
  print(f"{N-i}", end="")
  i += 1
print("\n")
```

```python
#5 Solicitar uma palavra ao usuário e dizer se a palavra é um palindromo, isto é pode ser lida igualmente da esquerda para a direita e vice-versa.
#Exemplos de palavras que são palindromos: ovo, radar, mirim, arara, reviver.
#Algumas dicas:
#palavra = "dados"
#N= len(palavra) # tamanho de uma palavra em caracteres
#print(N) # saída esperada: 5
#indexando caracteres de uma palavra:
#print(palavra[0]) # saida esperada: d
#print(palavra[1]) # saida esperada: a
#print(palavra[N-1]) # saida esperada:s
#print(palavra[-1]) # saida esperada: s

palavra= input("Digite uma palavra: ") #Variável para o usuário digitar a palavra
palavra = palavra.lower() #lower é para converter a palavra digitada para minusculo
palavra_invertida= "" #será utilizada para armazenar a palavra vazia

for i in range (len(palavra) -1, -1, -1):
  palavra_invertida += palavra[i]

palindromo= palavra_invertida == palavra

if palindromo:
  print("A palavra é um palíndromo!")

else:
  print("A palavra não é um palindromo")
```

```python
#6 Fazer um programa que solicita um número inteiro positivo N ao usuário e imprime o seguinte:
"""
N= 1:
 -
| |
 -

 N= 2:
 --
|  |
|  |
 --

N= 5:
 -----
|     |
|     |
|     |
|     |
|     |
|     |
 -----

"""
valores_N = [1, 2, 5] #as variaveis que precisa imprimir


for N in valores_N: # fazer a repetição para o N 1
    print(f"N = {N}:")
    print(' ' + '-' * N) #trazer a resposta na tela de um espaço e 1 traço
    for _ in range(N):
        print('|' + ' ' * N + '|') #trazer a resposta na tela com uma barra + um espaço e multiplicar o N e colocar mais uma barra

    # Imprime a linha inferior
    print(' ' + '-' * N)  #preencher um espaço +  a linha inferior e multiplicar o N para trazer a qtd desejada
    print()  # Add uma linha em branco
```
